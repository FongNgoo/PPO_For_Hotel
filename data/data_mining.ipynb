```python
import pandas as pd
import numpy as np
import datetime
```

```python
data_path = r"D:\Project\PPO_For_Hotel\data\hotel_bookings.csv"
df = pd.read_csv(data_path)
df
```

```python
df = df.loc[df["hotel"]=="Resort Hotel",:]
df.info()
```

```python
avg_adr = np.average(df["adr"])
print(avg_adr)
```

```python
month_map = {
    "January": 1, "February": 2, "March": 3, "April": 4,
    "May": 5, "June": 6, "July": 7, "August": 8,
    "September": 9, "October": 10, "November": 11, "December": 12
}

df["arrival_date_month"] = df["arrival_date_month"].map(month_map)
```

```python
df["arrival_date"] = pd.to_datetime(
    dict(
        year=df["arrival_date_year"],
        month=df["arrival_date_month"],
        day=df["arrival_date_day_of_month"]
    )
)

df["booking_date"] = df["arrival_date"] - pd.to_timedelta(df["lead_time"], unit="D")
time_series = df["booking_date"]
print(f"First date of data set is {min(time_series)} and last date of dataset is {max(time_series)}")
```

```python
data_path = r"D:\Project\PPO_For_Hotel\data\resort_hotel_data.csv"
df.to_csv(data_path)
```
